```python
# 03. Data visualization
# a. import the bank-data.csv file
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

bank_df =pd.read_csv('bank-data.csv')
```

```python
# b. Retrieve the first five records in the data set
bank_df.head()
```

```python
sns.pairplot(bank_df.select_dtypes(include=['float64','int64']))
plt.show()
```

```python
# d. Describe the data set
print(bank_df.describe(include='all'))
```

```python
# e. Plot the data set income Vs children
plt.figure(figsize=(8,5))
sns.scatterplot(x='income',y='children',data=bank_df)
plt.title("Income vs children")
plt.xlabel("Income")
plt.ylabel("Children")
plt.grid(True)
plt.show()
```

```python
# f. Construct the bar chart for region
plt.figure(figsize=(8,5))
sns.countplot(x='region',data=bank_df)
plt.xticks(rotation=45)
plt.title("Count of customer by region")
plt.show()
```

```python
#g. Construct the scatter plot for income and children and set the title as income vs children
plt.figure(figsize=(8,5))
sns.scatterplot(x='income',y='children',data=bank_df)
plt.title("Income vs Children")
plt.show()
```

```python
#h. Construct the histogram for children
plt.figure(figsize=(8,5))
sns.histplot(bank_df['children'],bins=10,kde=True)
plt.title("Distribution of children")
plt.xlabel("Number of chilren")
plt.show()
```

```python
# 04. Explanatory data analysis
# a. Import the libraries numpy, pandas, matplotlib.pyplot, seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# b. Import the data set bank-data.csv as data
data=pd.read_csv('bank-data.csv')
```

```python
# c. Check the data type
data.dtypes
```

```python
#d. Check the data shape
data.shape
```

```python
#e. Check the attribute names of the data set
data.columns.tolist()
```

```python
#f. Check the null values
print(data.isnull().sum())
```

```python
#g. Remove the unwanted attribute

print("Befor remove unwanted attribute in data set \n ##############################################\n")
print(data.columns.tolist())
if 'id' in data.columns:
    data.drop(columns=['id'],inplace=True)
print("After remove unwanted attribute in data set \n ##############################################\n")
print(data.columns.tolist())
```

```python
#h. Rename the Attribute name
print("Befor rename  attribute in data set \n ##############################################\n")
print(data.columns.tolist())

data.rename(columns={'income':'Income_Usd'},inplace=True)
print("After rename attribute in data set \n ##############################################\n")
print(data.columns.tolist())
```

```python
#i. Construct the plot for pep
plt.figure(figsize=(6,4))
sns.countplot(x='pep',data=data)
plt.title("Distribution of PEP")
plt.show()
```

```python
#j. Group the data sex and pep according to the count
grouped=data.groupby(['sex','pep']).size().reset_index(name='count')
print(grouped)
```

```python
#k. Construct the graph for pep vs sex
plt.figure(figsize=(6,4))
plt.title("PEP vs Sex")
sns.countplot(x='sex',hue='pep',data=data)
plt.show()

```

```python
#l. Find the count of the attribute children according to the category
children_count=data['children'].value_counts()
print("Children Count: \n",children_count)
```

```python
#m. Construct the graph for attribute children according to the category
plt.figure(figsize=(6,4))
sns.countplot(x='children',data=data)
plt.title("Number od customer by children count")
plt.show()
```

```python
#n. Construct the chart to illustrate pep, children, and sex of the customer based on the count
pivot = data.groupby(['pep', 'sex', 'children']).size().reset_index(name='count')
plt.figure(figsize=(10,6))
sns.catplot(x='children', y='count', hue='sex', col='pep', kind='bar', data=pivot)
plt.subplots_adjust(top=0.85)
plt.suptitle("PEP vs children vs sex -customer count")
plt.show()
```
