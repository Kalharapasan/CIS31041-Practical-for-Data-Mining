```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Company_ABC_HumanResource.csv')
```

```python
dataset.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(dataset['Position'],dataset['Salary'],color='blue',edgecolors='black')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
dc=dataset['Department'].value_counts()
plt.figure(figsize=(8,6))
dc.plot(kind='bar',color='blue',edgecolor='black')
plt.xlabel('Department')
plt.ylabel('Number of Employee')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(dataset['EmployerSatisfaction'],dataset['Salary'],color='blue')
plt.xlabel('Employer Satisfaction')
plt.ylabel('Salary')
plt.title('Employer Satisfaction vs Salary')
plt.tight_layout()
plt.show()
```

```python
dataset.isnull().sum()
```

```python
mode_value=dataset['MaritalStatus'].mode()[0]
dataset['MaritalStatus'].fillna(mode_value,inplace=True)
```

```python
mean_salary=dataset['Salary'].mean()

dataset['Salary'].fillna(mean_salary,inplace=True)
```
